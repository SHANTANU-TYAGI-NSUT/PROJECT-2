```python
# import python libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import csv file.

df = pd.read_csv("Student Data.csv")
print(df.head())
```

```python
# Displaying the number of rows and columns in the DataFrame.

df.shape
```

```python
# Generating descriptive statistics summarizing the central tendency,
# dispersion, and shape of the numerical columns in the DataFrame.

df.describe()
```

```python
# Providing a concise summary of the DataFrame, including the column names, data types, and memory usage.

df.info()
```

```python
# Checking the number of missing values in each column of the DataFrame.

df.isnull().sum()
```

# Drop unnamed column

```python
# Removing the column "Unnamed: 0" from the DataFrame.

df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
# Displaying the first few rows of the DataFrame to get an overview of the data.

df.head()
```

# Gender Distribution

```python
# Creating a count plot to visualize the distribution of gender in the DataFrame.

plt.figure(figsize= (5,7))
ax = sns.countplot(data = df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.title(" Gender Distribution")
plt.show()
```

# From the above chart we have analysed that  the number  

# of females in the data is more than the number of males.

```python
# Dropping rows with missing values in 'NrSiblings' or 'ReadingScore'.

df_cleaned = df.dropna(subset=['NrSiblings', 'ReadingScore'])

# Plotting using seaborn with cleaned data.

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='NrSiblings', y='ReadingScore')
plt.title('Relation between Number of Siblings and Reading Score')
plt.xlabel('Number of Siblings')
plt.ylabel('Reading Score')
plt.show()
```

# From the above plot, we analyzed the relationship between the number of 

# siblings and reading score.

```python
# Grouping the DataFrame by the "ParentEduc" column and calculating the mean math, reading, and writing scores for each group.

gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb)
```

```python
# Creating a heatmap to visualize the relationship between parent's education level and student's scores.

plt.figure(figsize= (6,6))
sns.heatmap(gb, annot=True)
plt.title(" Ralationship between Parent's Education and Student's Score")
plt.show()
```

# From the above chart we have concluded that the education

#  of the parents have a good impact on their scores.

```python
# Grouping the DataFrame by the "ParentMaritalStatus" column and calculating the mean math, 
# reading, and writing scores for each group.

gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb1)
```

```python
# Creating a heatmap to visualize the relationship between parent's marital status and student's scores.

plt.figure(figsize= (6,6))
sns.heatmap(gb1, annot=True)
plt.title(" Ralationship between Parent's Education and Student's Score")
plt.show()
```

# From the above chart we have concluded that there is 

# no/negligible impact on the student score due to their 

# parent's marrital status.

# Distribution of Ethnic Groups

```python
# Counting the number of entries for "group A" in the "EthnicGroup" column and printing the result.


groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
print(groupA)
```

```python
# Counting the number of entries for each ethnic group in the "EthnicGroup" column.

groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

# Creating a list of ethnic group labels and their corresponding counts.

l = ["group A", "group B", "group C" , "group D", "group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]] 

# Plotting a pie chart to visualize the distribution of ethnic groups.

plt.pie(mlist,labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")

plt.show()
```

```python
# Creating a count plot to visualize the number of students in each ethnic group and adding labels to the bars.

ax = sns.countplot(data = df, x = 'EthnicGroup')
ax.bar_label(ax.containers[0])
```

```python
# Printing the list of counts for each ethnic group.

print(mlist)
```

```python
# Dropping rows with missing values in 'NrSiblings'.

df_cleaned = df.dropna(subset=['NrSiblings'])

# Calculating the distribution of the number of siblings.

sibling_counts = df_cleaned['NrSiblings'].value_counts().sort_index()

# Plotting the pie chart.

plt.figure(figsize=(10, 10))
plt.pie(sibling_counts, labels=sibling_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(sibling_counts))))
plt.title('Distribution of Students by Number of Siblings')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# The above pie chart demonstrates that the most common number 

# of siblings among students is 2, indicating a tendency towards

# moderate family sizes.
